```python
import os
import csv
```

```python
def extract_time(data):
    return data.split(" ")[-1].rstrip("\n")

experiments = "Nodes,Agents,SmallTasks,MedTasks,BigTasks,Order,Cooperation,Balanced,Backup"
with os.scandir(".") as entries:
    for entry in entries:
        if entry.is_file():
            if entry.name.endswith("00.txt"):
                with open(entry.name, 'r') as f:
                    last_line = f.readlines()[-1]
                experiments += entry.name.split("-")[0] + "," + extract_time(last_line)

with open('experiments_results.txt', 'a') as the_file:
    the_file.write(experiments)
```
